```python
def Euler(y0,t0,dydt,h):
    '''Euler's method function'''
    return y0 + dydt(t0) * h
```

```python
def do_Euler(x0,h,n):
    ts = np.zeros(n)
    xs = np.zeros(n)
    xs[0] = x0
    ts[0] = 0
    for i in range(1,n-1,1):
        ts[i] = ts[i-1] + h
        xs[i] = Euler(xs[i-1],dxdt(ts[i-1]),h)
    
    return ts,xs
```
